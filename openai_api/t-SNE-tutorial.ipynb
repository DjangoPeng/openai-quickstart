```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
```

# Digit-image Example

```python
from sklearn import datasets

# load 64-dimensional data of 8x8 images
digits = datasets.load_digits()

# take the first 500 records
data = digits.data[:500]
labels = digits.target[:500]

data.shape
```

```python
# show one of the images
img = data[20].reshape(8,8)
plt.imshow(img)
```

```python
tsne = TSNE(n_components=2, perplexity=20)
data_2d = tsne.fit_transform(data)
```

```python
# scatter plt
# very well defined clusters with t-SNE
plt.scatter(data_2d[:,0],data_2d[:,1])
```

```python
# visualize and highlight 
target_ids = range(len(digits.target_names))

plt.figure(figsize=(10,8))
colors = ['r','g','b','c','m','y','k','pink','orange','purple']

for i, c, label in zip(target_ids, colors, digits.target_names):
    plt.scatter(data_2d[labels == i, 0], data_2d[labels == i, 1], c=c, label=label)
    pass

plt.legend()
plt.show()
```

# 3D Example

```python
# create two clusters of normal data
c1 = np.random.normal(2.0, scale=0.5, size=(50,3))

c2 = np.random.normal(0.0, scale=0.5, size=(50,3))
c2[:,2] = c2[:,2] + 3
c3 = np.random.normal(0.0, scale=0.5, size=(50,3))
c3[:,0] = c3[:,0] + 3
c4 = np.random.normal(0.0, scale=0.5, size=(50,3))
c4[:,0] = c4[:,0] - 2
c5 = np.random.normal(0.0, scale=0.5, size=(50,3))
c5[:,1] = c5[:,1] + 2

c6 = np.random.normal(-2.0, scale=0.5, size=(50,3))

data = np.concatenate((c1,c2,c3,c4,c5,c6),axis=0)
```

```python
# plt 2d: c6 is clearly a cluster, but other 5s are indistinguishable
fig = plt.figure(figsize=(14,4))
ax1 = fig.add_subplot(131)
ax1.scatter(data[:,0],data[:,1])
ax2 = fig.add_subplot(132)
ax2.scatter(data[:,1],data[:,2])
ax3 = fig.add_subplot(133)
ax3.scatter(data[:,2],data[:,0])
plt.show()
```

```python
# Create a 3d-figure
fig = plt.figure()

# Create a 3D axis
ax = fig.add_subplot(111, projection='3d')

# Create a scatter plot
ax.scatter(data[:,0], data[:,1], data[:,2])

# Set labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Show the plot
plt.show()
```

```python
tsne = TSNE(n_components=2, perplexity=15)
data_2d = tsne.fit_transform(data)
```

```python
# scatter plt data_3d
# very well defined clusters with t-SNE
plt.scatter(data_2d[:,0],data_2d[:,1])
```

# 2D Example

```python
# create two clusters of normal data
c1 = np.random.normal(5.0, scale=0.5, size=(50,2))
c2 = np.random.normal(-5.0, scale=0.5, size=(50,2))
c3 = np.random.normal(0.0, scale=0.5, size=(50,2))

data = np.concatenate((c1,c2,c3),axis=0)
```

```python
# scatter plt
plt.scatter(data[:,0],data[:,1])
```

```python
tsne = TSNE(n_components=2, perplexity=15)
data_2d = tsne.fit_transform(data)
```

```python
# scatter plt data_2d
# t-SNE exaggerates the cluster and don't read too much into where.
plt.scatter(data_2d[:,0],data_2d[:,1])
```
